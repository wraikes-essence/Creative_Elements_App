## Data Prep

```python
df_raw = df_raw[
    (
        (df_raw['Video_Length'] == ':06') |
        (df_raw['Video_Length'] == ':15') |
        (df_raw['Video_Length'] == ':30')
    ) &
    (df_raw['Publisher'] == 'YouTube')
]
```

```python
df = pd.get_dummies(df_raw, columns=
                    [
                        'Product',
                        'Region',
                        'Video_Length',
                        'Platform',
                        'Real_World',
                        'Story_Driven',
                        'Event', 
                        'Google_Logo_Upfront_Recode',
                        'Product_Logo_Upfront_Recode',
                        'Audio_Mention_Upfront_Recode',
                        'No_of_Visuals_Recode',
                        'Google_Logo_50',
                        'Product_Logo_50',
                        'Product_Shot_50',
                        'Text_on_End_Card',
                        'Demo',
                        'Front_Card',
                        'End_Card',
                        'Pop_Culture',
                        'Visually_Text_Only',
                        'Music',
                        'Voiceover',
                        'No_Product_Msgs_Recode'
                    ], drop_first=False
                   )
```

```python
df = df[[
    'Aided_Awareness',
    'Consideration_Intent',
    'Purchase_Intent_Top2Box',
        
    'Product_Achat', 
    'Product_Android OS', 
    'Product_Android Pay',
    'Product_Android Wear', 
    'Product_Chromebook', 
    'Product_Chromecast',
    'Product_Chromecast Audio', 
    'Product_Daydream View',
    'Product_Digital Skills', 
    'Product_Duo', 
    'Product_Ellen Show',
    'Product_Feed', 
    'Product_G Suite',
    'Product_GSA',
    'Product_Google Activate', 
    'Product_Google Assistant',
    'Product_Google Cloud', 
    'Product_Google Duo', 
    'Product_Google Express',
    'Product_Google Home', 
    'Product_Google Home Max',
    'Product_Google Home Mini', 
    'Product_Google Photos',
    'Product_Google Play', 
    'Product_Google Store',
    'Product_Google User Trust', 
    'Product_Nexus', 
    'Product_Pixel',
    'Product_Pixel 2', 
    'Product_Pixelbook', 
    'Product_Project Fi',
    'Product_Watercooler BCE', 
    'Product_YouTube Music',
    'Product_YouTube Music/Emerging Artists', 
    'Product_YouTube Red',
    'Product_YouTube TV',
    
    'Region_APAC',
    'Region_EMEA',
    'Region_North_America',
    'Platform_Mobile',
    'Platform_Desktop',
    'Video_Length_:06',
    'Video_Length_:15',
    'Video_Length_:30',
    
    'Real_World_animated',
    'Real_World_both',
    'Real_World_real world',
    'Story_Driven_no',
    'Story_Driven_yes',
    'Event_no',
    'Event_yes',
    'Google_Logo_Upfront_Recode_1+',
    'Google_Logo_Upfront_Recode_Zero',
    'Product_Logo_Upfront_Recode_1+',
    'Product_Logo_Upfront_Recode_Zero',
    'Audio_Mention_Upfront_Recode_1+',
    'Audio_Mention_Upfront_Recode_Zero',
    'No_of_Visuals_Recode_One+',
    'No_of_Visuals_Recode_Zero',
    'Google_Logo_50_no',
    'Google_Logo_50_yes',
    'Product_Logo_50_no',
    'Product_Logo_50_yes',
    'Product_Shot_50_no',
    'Product_Shot_50_yes',
    'Text_on_End_Card_no',
    'Text_on_End_Card_yes',
    'Demo_no',
    'Demo_yes',
    'Front_Card_no',
    'Front_Card_yes',
    'Pop_Culture_no',
    'Pop_Culture_yes',
    'Music_no',
    'Music_yes',
    'Voiceover_no',
    'Voiceover_yes',
    'No_Product_Msgs_Recode_One',
    'No_Product_Msgs_Recode_Two+',
    'No_Product_Msgs_Recode_Zero' 
]].dropna()

X = df.drop(axis=1, columns=['Aided_Awareness', 'Consideration_Intent', 'Purchase_Intent_Top2Box'])

y_aware = df['Aided_Awareness']
y_cons = df['Consideration_Intent']
y_purch = df['Purchase_Intent_Top2Box']
```

## Model KPI Rates

### Awareness

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y_aware, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

params = [{
    'n_estimators': [10, 50, 250],
    'max_depth': [1, 6, 10],
    'max_features': ['auto'] 
}]

rf_aware = RandomForestClassifier(random_state=42)
rf_grid_aware = GridSearchCV(rf_aware, params)
rf_grid_aware.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score

auc_aware = roc_auc_score(y_test, rf_grid_aware.best_estimator_.predict(X_test))
acc_aware = accuracy_score(y_test, rf_grid_aware.best_estimator_.predict(X_test))

auc_aware, acc_aware
```

```python
rf_aware = rf_grid_aware.best_estimator_
```

### Consideration

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_cons, test_size=0.2, random_state=42)
```

```python
params = [{
    'n_estimators': [10, 50, 250],
    'max_depth': [1, 6, 10],
    'max_features': ['auto'] 
}]

rf_cons = RandomForestClassifier(random_state=42)
rf_grid_cons = GridSearchCV(rf_cons, params)
rf_grid_cons.fit(X_train, y_train)
```

```python
auc_cons = roc_auc_score(y_test, rf_grid_cons.best_estimator_.predict(X_test))
acc_cons = accuracy_score(y_test, rf_grid_cons.best_estimator_.predict(X_test))

auc_cons, acc_cons
```

```python
rf_cons = rf_grid_cons.best_estimator_
```

### Purchase

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_purch, test_size=0.2, random_state=42)
```

```python
params = [{
    'n_estimators': [10, 50, 250],
    'max_depth': [1, 6, 10],
    'max_features': ['auto'] 
}]

rf_purch = RandomForestClassifier(random_state=42)
rf_grid_purch = GridSearchCV(rf_purch, params)
rf_grid_purch.fit(X_train, y_train)
```

```python
auc_purch = roc_auc_score(y_test, rf_grid_purch.best_estimator_.predict(X_test))
acc_purch = accuracy_score(y_test, rf_grid_purch.best_estimator_.predict(X_test))

auc_purch, acc_purch
```

```python
rf_purch = rf_grid_purch.best_estimator_
```
